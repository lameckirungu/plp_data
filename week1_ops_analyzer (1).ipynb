```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculate the efficiency percentage for an operational site.

    Parameters:
        actual_output (float): volume or units produced
        target_output (float): planned/target volume for the same period.
    Return:
        float: as a percentage
    """
    if target_output == 0:
        raise ValueError("Target cannot be zero")
    efficiency = (actual_output / target_output) * 100
    return round(efficiency, 2)
    
def get_operational_status(efficiency):
    """
    Classify an efficiency score into an operational status tier.

    Parameters:
        efficiency (float): as a percentage
    Returns:
        str
    """
    
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:
        return "Warning"
    else:
        return "Normal"

depot_network = [
    {
        "name": "Depot Alpha - Nairobi North",
        "actual_output": 18_450,
        "target_output": 20_000,
    },
    {
        "name": "Depot Bravo — Mombasa Port",
        "actual_output": 13_200,
        "target_output": 19_000
    },
    {
        "name": "Depot Charlie — Kisumu West",
        "actual_output": 22_750,
        "target_output": 24_000
    }
]
print(f" Network loaded: {len(depot_network)} depots registered.")

for site in depot_network:
    name = site["name"]
    actual = site["actual_output"]
    target = site["target_output"]

    efficiency = calculate_efficiency(actual, target)
    status = get_operational_status(efficiency)

    print(f"""
{icon}  {name}
   ├─ Actual Output  : {actual:,} KL
   ├─ Target Output  : {target:,} KL
   ├─ Efficiency     : {efficiency}%
   └─ Status         : {status}
""")
```
